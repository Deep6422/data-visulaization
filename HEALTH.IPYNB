# Visualization

```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
# Load dataset

# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("he.csv")
df.head(20)
```

# Line Graph
From this we can see that the graph is scattered meaning that there is no linera relationship meaning that  age dosent affect the SysBP it depends upon one's health itself.

```python
Age = [64, 21, 30, 35, 39, 68, 50, 23, 34, 36, 39, 25, 33, 58, 54, 68, 44, 54, 56, 60]
SysBP = [141, 109, 112, 106, 140, 115, 125, 111, 144, 113, 146, 108, 117, 143, 128, 116, 130, 117, 117, 106]
Age, SysBP = zip(*sorted(zip(Age, SysBP)))
plt.plot(Age, SysBP, color='green', marker='+', linestyle='dotted')

# title
plt.title("age vs SysBP")

# add a label to the y-axisYears
plt.ylabel("Age")
plt.xlabel("SysBP")
plt.show()
```

# Bar PLot 
Bar plot for SysBP of first 10 persons

```python



persons = [f"Per {i+1}" for i in range(10)]
SysBP = [141, 109, 112, 106, 140, 115, 125, 111, 144, 113]

# Create the bar plot
plt.bar(range(len(persons)), SysBP)

# Add title and labels
plt.title("SysBP of 10 People")
plt.ylabel("SysBP")
plt.xlabel("Persons")

plt.xticks(range(len(persons)), persons)

# Display the plot
plt.show()
```

# Pie Chart
This pie chart shows that 55% of the people need no treatment they are healthy but 45% need treatment it says they are not healthy

```python

df['indication'].value_counts().plot(kind='pie',autopct='%0.2f')
plt.title("Indication")
```

# Histogram
The graph shows how the ages of people with and without health problems are distributed. People without health problems tend to be older and have a wider age range compared to those with potential health issues.


```python

sns.kdeplot(df[df['indication'] == 0]['age'], label='No Indication')
sns.kdeplot(df[df['indication'] == 1]['age'], label='Indication')

plt.title('Age Distribution by Indication')
plt.xlabel('Age')
plt.ylabel('Density')

plt.legend()

plt.show()
```

# Heat Map
It shows the heat map between sex and indication here 0 means male and 1 means female for sex and for indication 0 means no health issues and one means health issues so form this heat map we can see that there is  most likely male to ahve health issue according to the data than female

```python

sns.heatmap(pd.crosstab(df['sex'],df['indication'])) 
```
